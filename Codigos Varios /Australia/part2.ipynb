```python
S=3
T=6
```

```python
nodos1=[k+1 for k in range(6)]
nodos1
```

```python
nodos2=[k for k in nodos1 if k!=S if k!=T  ]
nodos2
```

```python
from docplex.mp.model import Model
mdl=Model('ruta_mascorta')
```

```python
arcos=[(1,2),(1,3),(2,3),(2,4),(2,5),(3,5),(4,6),(5,4),(5,6)]
print(arcos)
```

```python
x=mdl.integer_var_dict(arcos,name='x')
```

```python
Costos={(1,2):2,(1,3):4,(2,3):1,(2,4):4,(2,5):2,(3,5):3,(4,6):2,(5,4):3,(5,6):2}
```

```python
x=mdl.integer_var_dict(arcos,name='x')
```

```python
mdl.minimize(mdl.sum(x[i,j]*Costos[i,j] for i,j in arcos))
```

```python
#sale del nodo inicial
mdl.add_constraint(mdl.sum(x[i,j] for i,j in arcos if i==S)== 1)
```

```python
#ingresa al nodo final 
mdl.add_constraint(mdl.sum(x[i,j] for i,j in arcos if j==T)== 1)
```

```python
print(mdl.export_to_string())
```

```python
for k in nodos2: 
    mdl.add_constraint(mdl.sum(x[i,j] for i,j in arcos if i==k) -mdl.sum(x[j,i] for j,i in arcos if i==k)    == 0)
```

```python
solucion=mdl.solve(log_output=True)
```
